# **Exploratory Data Analysis (EDA) with Visualization**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")


data=pd.read_csv('train.csv')
```

### **1. Gender Distribution**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=data, palette='coolwarm')
plt.title('Gender Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=14, color='blue')
plt.ylabel('Count', fontsize=14, color='green')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### **2 .Age Distribution**

```python
sns.set_style("whitegrid")  


plt.figure(figsize=(10, 6))

sns.histplot(data['Age'], kde=True, color='skyblue', bins=15, linewidth=1.5, 
             line_kws={"linewidth": 2, "color": "darkorange"})  


plt.title('Age Distribution', fontsize=18, fontweight='bold', color='black')

plt.xlabel('Age', fontsize=14, fontweight='bold', color='black')
plt.ylabel('Frequency', fontsize=14, fontweight='bold', color='black')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.5)

plt.tight_layout()
plt.show()

```

### **3. Study Satisfaction vs. Depression**

```python
sns.set_style("whitegrid") 
plt.figure(figsize=(8, 5))


sns.boxplot(x='Depression', y='Study Satisfaction', data=data, palette='Set1')


plt.title('Study Satisfaction vs Depression', fontsize=18, fontweight='bold', color='darkslategray')

plt.xlabel('Depression Status', fontsize=14, fontweight='bold', color='black')
plt.ylabel('Study Satisfaction', fontsize=14, fontweight='bold', color='black')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.show()

```

### **4. Work Pressure vs. Financial Stress**

```python
sns.set_style("whitegrid") 


plt.figure(figsize=(8, 5))


sns.scatterplot(x='Work Pressure', y='Financial Stress', data=data, hue='Gender', palette='deep', s=100, edgecolor='black', alpha=0.7)

plt.title('Work Pressure vs Financial Stress', fontsize=18, fontweight='bold', color='darkslategray')


plt.xlabel('Work Pressure', fontsize=14, fontweight='bold', color='black')
plt.ylabel('Financial Stress', fontsize=14, fontweight='bold', color='black')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


plt.legend(title='Gender', title_fontsize=12, fontsize=12, loc='upper right')

plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()

plt.show()

```

### **5. Gender vs. Depression**


```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Depression', data=data, palette='Set1')
plt.title('Gender vs Depression', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Depression', loc='upper right')
plt.show()
```
